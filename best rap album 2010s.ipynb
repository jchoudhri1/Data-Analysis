```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/jchoudhri1/Datasets/album_data.csv")

# Drop first row , which just repeats the column names
df.drop(df.index[0], inplace = True)
```

```python
# Fill NaN values with 50
df = df.fillna(50)
df.head()

# Change scores from str to int
df[["DJBooth.net", "Paste Magazine", "Consequence of Sound", "BrooklynVegan", "HipHopGoldenAge", "TheFreshCommitte", "Genius"]] = df[["DJBooth.net", "Paste Magazine", "Consequence of Sound", "BrooklynVegan", "HipHopGoldenAge", "TheFreshCommitte", "Genius"]].astype(int)

#Create new column showing average score from 7 publications
df["Mean Score"] = (df["DJBooth.net"] + df["Paste Magazine"] + df["Consequence of Sound"] + df["BrooklynVegan"] + df["HipHopGoldenAge"] + df["TheFreshCommitte"] + df["Genius"])/7
df
```

```python
results = df.groupby("Album").sum()
results
```

```python
albums = [album for album, df in df.groupby('Album')]
plt.bar(albums, results ['Mean Score'])
plt.xticks(albums, rotation = 'vertical', size = 8)
plt.ylabel('Mean Score')
plt.xlabel('Album Name')
plt.show()

```
